```python
import glob
import os
import re
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import zipfile

print("TensorFlow version:", tf.__version__)

# 0. zip 파일 압축 해제 
zip_files = glob.glob('/content/*.zip')
if zip_files:
    print(f"찾은 zip 파일: {zip_files}")
    for zip_file in zip_files:
        if 'lyric' in zip_file.lower():
            print(f"압축 해제 중: {zip_file}")
            with zipfile.ZipFile(zip_file, 'r') as zip_ref:
                zip_ref.extractall('/content/')
            print("압축 해제 완료!")
            break

# 1. 데이터 읽어오기 
txt_file_path = os.getenv('HOME')+'/aiffel/lyricist/data/lyrics/*'

txt_list = glob.glob(txt_file_path)

raw_corpus = []

# 여러개의 txt 파일을 모두 읽어서 raw_corpus 에 담습니다.
for txt_file in txt_list:
    with open(txt_file, "r") as f:
        raw = f.read().splitlines()
        raw_corpus.extend(raw)

print("데이터 크기:", len(raw_corpus))
print("Examples:\n", raw_corpus[:3])

# 2. 데이터 정제 (평가기준 1: 특수문자 제거, 토크나이저 생성, 패딩 처리)
def preprocess_sentence(sentence):
    sentence = sentence.lower().strip()  # 소문자로 바꾸고 양쪽 공백을 삭제
    sentence = re.sub(r"([?.!,¿])", r" \1 ", sentence)  # 패턴의 특수문자를 만나면 특수문자 양쪽에 공백을 추가
    sentence = re.sub(r'[" "]+', " ", sentence)  # 공백 패턴을 만나면 스페이스 1개로 치환
    sentence = re.sub(r"[^a-zA-Z?.!,¿]+", " ", sentence)  # a-zA-Z?.!,¿ 패턴을 제외한 모든 문자를 스페이스 1개로 치환
    sentence = sentence.strip()
    return sentence

# 수정: 너무 긴 문장만 제거 (40개 토큰 이하로 늘림)
corpus = []

for sentence in raw_corpus:
    preprocessed_sentence = preprocess_sentence(sentence)
    if len(preprocessed_sentence) == 0:
        continue
    # 수정: 40개 토큰 이하로 늘림 (더 많은 데이터 확보)
    if len(preprocessed_sentence.split()) <= 40:
        corpus.append("<start> " + preprocessed_sentence + " <end>")

print("전처리 후 데이터 개수:", len(corpus))
print("전처리 후 Examples:\n", corpus[:3])

# 3. 평가기준 1: 토크나이저 생성
# 수정: 단어장 크기 12,000 이상 
def tokenize(corpus):
    tokenizer = Tokenizer(num_words=12000,  # 수정: 12,000 단어로 제한
                          filters='#$%&()*+-./:;<=>@[\\]^_`{|}~\t\n',  # 수정: ! , ? 를 필터에서 제외
                          oov_token="<unk>")  # Out-Of-Vocabulary 토큰
    tokenizer.fit_on_texts(corpus)
    
    tensor = tokenizer.texts_to_sequences(corpus)
    
    # 패딩 처리 (평가기준 1)
    tensor = pad_sequences(tensor, padding='post')
    
    print("Total words:", len(tokenizer.word_index))
    print("Tensor shape:", tensor.shape)
    
    return tensor, tokenizer

tensor, tokenizer = tokenize(corpus)

# 4. 입력과 레이블 분리
# 소스 문장 (입력): <start> ~ 마지막에서 두 번째 토큰까지
# 타겟 문장 (레이블): 두 번째 토큰 ~ <end>까지
src_input = tensor[:, :-1]
tgt_input = tensor[:, 1:]

print("Source input shape:", src_input.shape)
print("Target input shape:", tgt_input.shape)

# 5. 평가 데이터셋 분리 (수정: 10%로 줄여서 학습 데이터 더 확보)
from sklearn.model_selection import train_test_split

enc_train, enc_val, dec_train, dec_val = train_test_split(src_input, 
                                                            tgt_input,
                                                            test_size=0.2,  
                                                            random_state=42)

print("Train 데이터 크기:", enc_train.shape)
print("Validation 데이터 크기:", enc_val.shape)

# 6. 데이터셋 구성
BUFFER_SIZE = len(enc_train)
BATCH_SIZE = 256  # 수정: 배치 크기 조정 (예시 코드 참고)
steps_per_epoch = len(enc_train) // BATCH_SIZE

VOCAB_SIZE = len(tokenizer.word_index) + 1

# Dataset 생성
dataset = tf.data.Dataset.from_tensor_slices((enc_train, dec_train)).shuffle(BUFFER_SIZE)
dataset = dataset.batch(BATCH_SIZE, drop_remainder=True)

# 7. 평가기준 3: 안정적 학습을 위한 모델 구성
class TextGenerator(tf.keras.Model):
    def __init__(self, vocab_size, embedding_size, hidden_size):
        super(TextGenerator, self).__init__()
        
        self.embedding = tf.keras.layers.Embedding(vocab_size, embedding_size)
        # 수정: LSTM 레이어 2개 추가 (안정적 학습)
        self.lstm_1 = tf.keras.layers.LSTM(hidden_size, return_sequences=True)
        self.lstm_2 = tf.keras.layers.LSTM(hidden_size, return_sequences=True)
        # 추가: Dropout으로 과적합 방지
        self.dropout = tf.keras.layers.Dropout(0.2)
        self.linear = tf.keras.layers.Dense(vocab_size)
        
    def call(self, x):
        out = self.embedding(x)
        out = self.lstm_1(out)
        out = self.dropout(out)
        out = self.lstm_2(out)
        out = self.dropout(out)
        out = self.linear(out)
        
        return out

# 수정: Embedding Size와 Hidden Size 조절 (예시 코드 권장사항)
embedding_size = 256  # 수정: 임베딩 크기 증가
hidden_size = 1024    # 수정: Hidden size 증가

lyricist = TextGenerator(vocab_size=VOCAB_SIZE, 
                         embedding_size=embedding_size, 
                         hidden_size=hidden_size)  # 수정: 변수명을 lyricist로 변경

# 8. Loss 함수 
loss = tf.keras.losses.SparseCategoricalCrossentropy(
    from_logits=True, reduction='none')

# 9. 학습 과정 정의
# 수정: learning rate 조정
lyricist.compile(loss=loss, 
                 optimizer=tf.keras.optimizers.Adam(learning_rate=0.001))

# 10. 평가기준 3: 안정적 학습을 위한 콜백 설정
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

# EarlyStopping: validation loss 2.2 이하 목표
early_stopping = EarlyStopping(
    monitor='val_loss',
    patience=5,  # 5 에포크 동안 개선 없으면 중단
    restore_best_weights=True,
    verbose=1
)

# 수정: 서브클래스 모델은 SavedModel 형식으로 저장
checkpoint = ModelCheckpoint(
    'best_lyrics_model',  # 수정: .h5 제거 (SavedModel 형식)
    monitor='val_loss',
    save_best_only=True,
    save_format='tf',  # 추가: TensorFlow SavedModel 형식
    verbose=1
)

# 11. 모델 학습 (수정: 에포크 줄이고 데이터로 보완)
history = lyricist.fit(dataset, 
                       epochs=15,  # 수정: 30 → 15로 줄임 (데이터가 충분하므로)
                       validation_data=(enc_val, dec_val),
                       callbacks=[early_stopping, checkpoint],
                       verbose=1)

# 12. 학습 결과 시각화
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.axhline(y=2.2, color='r', linestyle='--', label='Target (2.2)')  # 목표선
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.title('Model Loss')
plt.grid(True)

plt.tight_layout()
plt.show()

print(f"\n최종 Validation Loss: {min(history.history['val_loss']):.4f}")
print(f"목표 달성 여부 (2.2 이하): {min(history.history['val_loss']) <= 2.2}")

```

```python
# 결과가 너무 단순해서 다시 시도

import glob
import os
import re
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split

print("TensorFlow version:", tf.__version__)

# 1. 데이터 읽어오기
txt_file_path = os.getenv('HOME')+'/aiffel/lyricist/data/lyrics/*'
txt_list = glob.glob(txt_file_path)

raw_corpus = []
for txt_file in txt_list:
    with open(txt_file, "r") as f:
        raw_corpus.extend(f.read().splitlines())

print("원본 데이터 수:", len(raw_corpus))
print("Examples:", raw_corpus[:3])

# 2. 전처리 함수
def preprocess_sentence(sentence):
    sentence = sentence.lower().strip()
    sentence = re.sub(r"([?.!,¿])", r" \1 ", sentence)
    sentence = re.sub(r'[" "]+', " ", sentence)
    sentence = re.sub(r"[^a-zA-Z?.!,¿]+", " ", sentence)
    sentence = sentence.strip()
    sentence = ' ' + sentence + ' '  # 예시 코드와 동일
    return sentence

# 3. 데이터 전처리
corpus = list(map(preprocess_sentence, raw_corpus))
print("전처리 후 데이터 개수:", len(corpus))
print("Examples:", corpus[:3])

# 4. 토크나이저 및 Tensor 변환
def tokenize(corpus, max_len=40):  # max_len 40으로 늘려 학습 데이터 확보
    tokenizer = Tokenizer(
        num_words=12000,
        filters='',
        oov_token="<unk>"
    )
    tokenizer.fit_on_texts(corpus)
    tensor = tokenizer.texts_to_sequences(corpus)
    tensor = pad_sequences(tensor, padding='post', maxlen=max_len)
    print("Tensor shape:", tensor.shape)
    print("단어 사전 크기:", len(tokenizer.word_index))
    return tensor, tokenizer

tensor, tokenizer = tokenize(corpus, max_len=40)

# 5. 입력과 레이블 분리
enc_inputs = tensor[:, :-1]
dec_targets = tensor[:, 1:]

# 6. train/validation split (20% 평가)
enc_train, enc_val, dec_train, dec_val = train_test_split(
    enc_inputs,
    dec_targets,
    test_size=0.2,
    random_state=42,
    shuffle=True
)

print("Train size:", enc_train.shape)
print("Validation size:", enc_val.shape)

# 7. Dataset 구성
BATCH_SIZE = 256  # 학습 속도 고려
BUFFER_SIZE = len(enc_train)

train_dataset = tf.data.Dataset.from_tensor_slices((enc_train, dec_train))
train_dataset = train_dataset.shuffle(BUFFER_SIZE).batch(BATCH_SIZE, drop_remainder=True)

val_dataset = tf.data.Dataset.from_tensor_slices((enc_val, dec_val))
val_dataset = val_dataset.batch(BATCH_SIZE, drop_remainder=True)

# 8. 모델 구성
class TextGenerator(tf.keras.Model):
    def __init__(self, vocab_size, embedding_size, hidden_size):
        super().__init__()
        self.embedding = tf.keras.layers.Embedding(vocab_size, embedding_size)
        self.rnn_1 = tf.keras.layers.LSTM(hidden_size, return_sequences=True)
        self.rnn_2 = tf.keras.layers.LSTM(hidden_size, return_sequences=True)
        self.dropout = tf.keras.layers.Dropout(0.2)  # 과적합 방지
        self.linear = tf.keras.layers.Dense(vocab_size)

    def call(self, x):
        out = self.embedding(x)
        out = self.rnn_1(out)
        out = self.dropout(out)
        out = self.rnn_2(out)
        out = self.dropout(out)
        out = self.linear(out)
        return out

embedding_size = 256
hidden_size = 1024
VOCAB_SIZE = tokenizer.num_words + 1

model = TextGenerator(VOCAB_SIZE, embedding_size, hidden_size)

# 9. 학습 설정
loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True, reduction='none')
optimizer = tf.keras.optimizers.Adam(learning_rate=0.001)

model.compile(loss=loss, optimizer=optimizer)

# 10. EarlyStopping 및 Checkpoint
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

early_stopping = EarlyStopping(
    monitor='val_loss',
    patience=2,             # 2 epoch 동안 개선 없으면 종료
    restore_best_weights=True,
    verbose=1,
    baseline=2.0            # val_loss 2.0 이하 시 목표 달성으로 간주
)

checkpoint = ModelCheckpoint(
    'best_lyrics_model',
    monitor='val_loss',
    save_best_only=True,
    save_format='tf',
    verbose=1
)

# 11. 학습
history = model.fit(
    train_dataset,
    validation_data=val_dataset,
    epochs=10,             # 충분히 학습 가능
    callbacks=[early_stopping, checkpoint],
    verbose=1
)

# 12. 텍스트 생성 함수
def generate_text(model, tokenizer, init_sentence="<start>", max_len=20):
    test_input = tokenizer.texts_to_sequences([init_sentence])
    test_tensor = tf.convert_to_tensor(test_input)
    end_token = tokenizer.word_index.get("<end>", None)

    for _ in range(max_len):
        predictions = model(test_tensor)
        predicted_id = tf.argmax(predictions[:, -1:, :], axis=-1, output_type=tf.int32)
        if end_token is not None and predicted_id == end_token:
            break
        test_tensor = tf.concat([test_tensor, predicted_id], axis=-1)

    generated_sequence = test_tensor.numpy()[0]
    generated_text = tokenizer.sequences_to_texts([generated_sequence])[0]
    return generated_text

# 13. 테스트 텍스트 생성
test_sentences = ["<start> i love", "<start> you are", "<start> we can"]
for s in test_sentences:
    print(f"\n입력: {s}")
    print("생성:", generate_text(model, tokenizer, init_sentence=s, max_len=20))


```

```python
import glob
import os
import re
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split

print("TensorFlow version:", tf.__version__)

# 1. 데이터 읽어오기
txt_file_path = os.getenv('HOME')+'/aiffel/lyricist/data/lyrics/*'
txt_list = glob.glob(txt_file_path)

raw_corpus = []
for txt_file in txt_list:
    with open(txt_file, "r") as f:
        raw_corpus.extend(f.read().splitlines())

print("원본 데이터 수:", len(raw_corpus))
print("Examples:\n", raw_corpus[:3])

# 2. 데이터 전처리
def preprocess_sentence(sentence):
    sentence = sentence.lower().strip()
    sentence = re.sub(r"([?.!,¿])", r" \1 ", sentence)
    sentence = re.sub(r'[" "]+', " ", sentence)
    sentence = re.sub(r"[^a-zA-Z?.!,¿]+", " ", sentence)
    sentence = sentence.strip()
    return ' ' + sentence + ' '

corpus = [preprocess_sentence(s) for s in raw_corpus]

# 3. 토크나이저 생성
def tokenize(corpus, max_len=15):
    tokenizer = Tokenizer(num_words=12000, filters=' ', oov_token="")
    tokenizer.fit_on_texts(corpus)
    tensor = tokenizer.texts_to_sequences(corpus)
    tensor = pad_sequences(tensor, padding='post', maxlen=max_len)
    print("Tensor shape:", tensor.shape)
    return tensor, tokenizer

tensor, tokenizer = tokenize(corpus, max_len=15)

# 4. 입력과 레이블 분리
enc_inputs = tensor[:, :-1]
dec_targets = tensor[:, 1:]

# 5. 학습/평가 데이터 분리
enc_train, enc_val, dec_train, dec_val = train_test_split(
    enc_inputs, dec_targets, test_size=0.2, random_state=42, shuffle=True
)

print("Train size:", enc_train.shape)
print("Validation size:", enc_val.shape)

# 6. Dataset 구성
BATCH_SIZE = 64
BUFFER_SIZE = len(enc_train)

train_dataset = tf.data.Dataset.from_tensor_slices((enc_train, dec_train)).shuffle(BUFFER_SIZE).batch(BATCH_SIZE, drop_remainder=True)
val_dataset = tf.data.Dataset.from_tensor_slices((enc_val, dec_val)).batch(BATCH_SIZE, drop_remainder=True)

# 7. 모델 구성
class TextGenerator(tf.keras.Model):
    def __init__(self, vocab_size, embedding_size=256, hidden_size=1024):
        super().__init__()
        self.embedding = tf.keras.layers.Embedding(vocab_size, embedding_size)
        self.rnn_1 = tf.keras.layers.LSTM(hidden_size, return_sequences=True)
        self.rnn_2 = tf.keras.layers.LSTM(hidden_size, return_sequences=True)
        self.linear = tf.keras.layers.Dense(vocab_size)
    def call(self, x):
        out = self.embedding(x)
        out = self.rnn_1(out)
        out = self.rnn_2(out)
        out = self.linear(out)
        return out

VOCAB_SIZE = tokenizer.num_words + 1
model = TextGenerator(VOCAB_SIZE)

# 8. 학습 설정
loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True, reduction='none')
optimizer = tf.keras.optimizers.Adam(learning_rate=0.001)
model.compile(loss=loss, optimizer=optimizer)

# 9. 학습
history = model.fit(train_dataset, validation_data=val_dataset, epochs=5)

# 10. 텍스트 생성 함수
def generate_text(model, tokenizer, init_sentence="<start>", max_len=20):
    test_tensor = tokenizer.texts_to_sequences([init_sentence])
    test_tensor = pad_sequences(test_tensor, maxlen=15, padding='post')
    test_tensor = tf.convert_to_tensor(test_tensor)
    result = init_sentence
    for _ in range(max_len):
        predictions = model(test_tensor)
        predicted_id = tf.argmax(predictions[:, -1, :], axis=-1).numpy()[0]
        if predicted_id == 0:  # pad
            break
        result += ' ' + tokenizer.index_word.get(predicted_id, '')
        test_tensor = tf.concat([test_tensor, [[predicted_id]]], axis=-1)
    return result

# 테스트
test_sentences = ["<start> i love", "<start> you are", "<start> we can"]
for s in test_sentences:
    print(generate_text(model, tokenizer, init_sentence=s, max_len=20))
```
